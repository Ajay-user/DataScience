```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import lars_path
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
```

## Dataset

```python
X, y = datasets.load_diabetes(return_X_y=True)

print('Shape of X :',X.shape)
```

## Scale the data

```python
X_scaled = StandardScaler().fit_transform(X)
```

## Regularization path

lasso path  returns 

* `alphas` -- array-like of shape (n_alphas + 1,) <br>
Maximum of covariances (in absolute value) at each iteration. n_alphas is either max_iter, n_features or the number of nodes in the path with alpha >= alpha_min, whichever is smaller.

* `active` -- array-like of shape (n_alphas,)<br>
Indices of active variables at the end of the path.

* `coefs`  -- array-like of shape (n_features, n_alphas + 1)<br>
Coefficients along the path.

* `n_iter`  --  int<br>
Number of iterations run. Returned only if return_n_iter is set to True.

```python
alphas, active, coefs = lars_path(X_scaled, y, method='lasso')
```

```python
print('Number of Features', X.shape[1])
print('Shape of alphas',alphas.shape)
print('Shape of coefs',coefs.shape)
```

```python
plt.figure(figsize=(10,8))
for i in range(10):
    plt.plot(alphas, coefs.T[:, i], label=datasets.load_diabetes().feature_names[i])
plt.legend();
plt.xlabel('Alphas');
plt.ylabel("Coefficients");
```
