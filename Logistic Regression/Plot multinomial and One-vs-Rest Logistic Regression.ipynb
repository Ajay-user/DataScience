```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.linear_model import LogisticRegression
from sklearn.inspection import DecisionBoundaryDisplay
```

## make 3-class dataset for classification

```python

centers = [[-5, 0], [0, 1.5], [5, -1]]

X, y = make_blobs(n_samples=1000, centers=centers, random_state=40)
```

## One v/s Rest

```python
lr = LogisticRegression(solver='sag', multi_class='ovr', max_iter=150).fit(X,y)

# plot decision boundary
DecisionBoundaryDisplay.from_estimator(lr, X, response_method='predict');

for i in range(3):
    idx = y==i
    plt.scatter(X[idx,0], X[idx, 1], edgecolors='k')
```

## Multinomial

```python
lr = LogisticRegression(solver='sag', multi_class='multinomial', max_iter=150).fit(X,y)

# plot decision boundary
DecisionBoundaryDisplay.from_estimator(lr, X, response_method='predict');

for i in range(3):
    idx = y==i
    plt.scatter(X[idx,0], X[idx, 1], edgecolors='k')
```
