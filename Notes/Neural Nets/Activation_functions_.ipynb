# Activation functions
* ### Linear activation function (pass-through).
* ### Threshold function
* ### Sigmoid-function
* ### Tan Hyperbolic (tanH)
* ### Rectified Linear Unit (ReLU)


# Create dummy data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf

x = tf.cast(tf.constant(np.arange(-10,10)),dtype=tf.dtypes.float32)

x
```

```python
## ploting the dummy data

plt.plot(x);
```

# Create a utility function for ploting input x and output after applying activation function

```python
def plot_input_and_output(inp,out,function_name='activation function'):
  plt.figure(figsize=(15,5))

  plt.subplot(1,2,1)
  plt.plot(inp)
  plt.title('Input x',fontsize=15)

  plt.subplot(1,2,2)
  plt.plot(out)
  plt.title('Output after applying {0}'.format(function_name),fontsize=15)
```

# Linear activation function (pass-through).

Returns the input, unmodified

```python
out = tf.keras.activations.linear(x)

# plot the input and output
plot_input_and_output(x,out,'linear activation function')
```

# Threshold Activation function

fucntion outputs zero if x is less than threshold

otherwise if x is greater than or equal to threshold function outputs one

> Threshold function is a boolean function

> Threshold is th simplest activation 

> We can use Threshold function for simple cases like AND operation, OR operation etc...

> Threshold function is not differentiable

> Tuning a model requires differentiation # Gradient descent 

```python
# lets create a basic threshold function

def threshold_fn(input,threshold):
  output=[]
  for i in range(len(input)):
    if input[i]>threshold:
      output.append(1)
    else:
      output.append(0)
  return tf.constant(output,dtype=tf.dtypes.float32)
      

#set threshold
thres = 0
out = threshold_fn(x,thres)

# plot the input and output
plot_input_and_output(x, out, 'Threshold Activation Function')
```

# Sigmoid Activation function

**Sigmoid activation function, `sigmoid(x) = 1 / (1 + exp(-x))`**.

Threshold function is not differentiable , training a model using treshold activation leads to problems , so we use `Sigmoid` as activation fuction

> Sigmoid Activation Function is Differentiable

> Output of Sigmoid ranges from 0 to 1

> The function is monotonic while its derivative is not monotonic.

> Used in Feed-Forward-Neural-Nets.

```python
out = tf.keras.activations.sigmoid(x)

#plot input and output
plot_input_and_output(x,out,'Sigmoid Activation Function')
```

# Tan Hyperbolic (tanH)
Hyperbolic tangent activation function.
**`tanh(x) = sinh(x)/cosh(x) = ((exp(x) - exp(-x))/(exp(x) + exp(-x)))`**.

> Differentiable

> Output ranges from -1 to 1

> The function is monotonic while its derivative is not monotonic.

> The negative inputs will be mapped negative and the zero inputs will be mapped near zero 

> Used in Feed-Forward-Neural-Nets.

```python
out = tf.keras.activations.tanh(x)

# plot input and output
plot_input_and_output(x,out,'Tan Hyperbolic Activation Function')
```

# Rectified Linear Unit (ReLU)

> In modern neural nets using sigmoid or tanh causes **`Gradient Vanishing Problem`** this can be remedied by using ReLU activation function

> With default values, this returns the standard ReLU activation: **max(x, 0)**, the element-wise maximum of 0 and the input tensor.

> The function and its derivative both are monotonic.

> Modifying default parameters allows you to use non-zero thresholds, change the max value of the activation, and to use a non-zero multiple of the input for values below the threshold.

> Range: [ 0 to infinity)

> **ReLU is more advantages than Sigmoid in modern deep nets**

```python
out = tf.keras.activations.relu(x)

# plot input and output
plot_input_and_output(x,out,'ReLU Activation Function')
```
